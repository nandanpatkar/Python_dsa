```python
def kadane(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    for i in range( len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        print(f"Current sum at index {i}: {current_sum}")
        max_sum = max(max_sum, current_sum)
        print(f"Max sum so far: {max_sum}")

    return max_sum


arr = [-2,1,-3,4,-1,2,1,-5,4]
print(kadane(arr))
```

# Bruitforce


```python

def kadan_bruitforce(arr):
    max_sum = arr[0]

    for i in range(len(arr)):
        current_sum = 0
        for j in range(i ,len(arr)):
            current_sum = current_sum + arr[j]
            if current_sum > max_sum:
                max_sum = current_sum
    return max_sum

arr = [-2,1,-3,4,-1,2,1,-5,4]
kadan_bruitforce(arr)
```
