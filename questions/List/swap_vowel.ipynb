```python

def swap_vowel(s):
    s = list(s)
    vowel = "aeiou"
    left = 0 
    right = len(s)-1 

    while left< right :
        if s[left] not in vowel:
            left = left +1 
        elif s[right]not in vowel:
            right = right - 1
        else :
            s[left], s[right] = s[right],s[left]
            left = left +1 
            right = right - 1
    result = "".join(s)
    return result

s = "pretestpassed"
swap_vowel(s)
```
