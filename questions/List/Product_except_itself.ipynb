```python
def sum_a(arr):
    n = len(arr)
    answer = [1] * n
    prefix = 1
    for i in range(n):
        answer[i] = prefix
        prefix *= arr[i]
    
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= arr[i]
    
    return answer

arr = [1, 2, 3, 4, 5]
print(sum_a(arr))


// [120, 60, 40, 30, 24]

```
