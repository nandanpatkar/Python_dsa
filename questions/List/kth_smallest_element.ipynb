# the code for kth largest and kth smalles is almost same just chnage the sign pf "i to -i" and 'heap[0] to -heap[0]"

```python


import heapq

def kth_smallest(arr,k):
    heap = []

    for i in arr:
            heapq.heappush(heap,-i)
            if len(heap)> k :
                heapq.heappop(heap)
    return -heap[0]

k = 3
arr = [3,2,1,5,6,4]
kth_smallest(arr,k)
       
```
