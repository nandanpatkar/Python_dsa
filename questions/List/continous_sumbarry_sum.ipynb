```python
def subarray_sum(arr,target):
    n = len(arr)
    maxi = 0

    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += arr[j]
            if current_sum == target:
                return arr[i:j+1]
    return []

arr = [1,2,3,7,5]
target = 3

subarray_sum(arr,target)
```
