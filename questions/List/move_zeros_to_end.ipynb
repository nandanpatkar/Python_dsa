```python
def optimal(arr):

    left = 0 
    for right in range(len(arr)):
        if  arr[right] !=0:
            arr[left],arr[right] = arr[right],arr[left]
            left = left + 1

    return arr
arr = [1,0,2,3,0,4]
optimal(arr)
```
