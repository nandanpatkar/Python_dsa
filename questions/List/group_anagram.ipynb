```python
words = ["eat","tea","tan","ate","nat","bat"]
def group_anagrams(words):
	group_anagram = {}
	for word in words:
		key = "".join(sorted(word))
		# print(key)
		if key not in group_anagram:
			group_anagram[key] = []
		group_anagram[key].append(word)
	return list(group_anagram.values())
```

```python
group_anagrams(words)
```
