# without using extra space ( store last element and then reverse loop and shift element )

```python

def rotate_by_1(arr):
    n = len(arr)
    last = arr[n-1]
    for i in range(len(arr)-1,0,-1):
        arr[i] = arr[i-1]
    arr[0]=last
    
arr = [2,4,6,8,10]  
rotate_by_1(arr)
 
```
