Given an array and a target sum, find all pairs whose sum equals the target.

```python
arr = [2,4,3,5,7,8]
target = 7



def pair_sum(arr):

    n = len(arr)
    pairs = []
    for i in range(n):
        for j in range(i+1 , n):
            if arr[i] + arr[j] == target:
              pairs.append((arr[i],arr[j]))
    return pairs

            
```

```python
pair_sum(arr)
```

# Using Two Pointer if array is Sorted 


```python
def pair_sum_two_pointer(arr,target):
    arr.sort()
    pairs = []
    left = 0
    right = len(arr)-1

    while left < right:
        sum = arr[left] + arr[right]

        if sum == target:
            pairs.append((arr[left],arr[right]))
            left += 1
            right -= 1
        elif sum < target:
            left = left + 1
        else :
            right = right - 1
    return pairs        


arr = [2,4,3,5,7,8]
target = 7    
pair_sum_two_pointer(arr,target)



```
