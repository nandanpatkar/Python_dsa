```python
arr1 = [ 1,2,3,4]
arr2 = [ 3,4,5,6]
```

```python
new_array = []
for i in range(len(arr1)):
    for j in range(len(arr2)):
        if arr1[i] != arr2[j] and arr2[j] not in new_array:
            new_array.append(arr1[i])
            new_array.append(arr1[j]) 
    
```
