# Given an array and a target sum, find a continuous subarray whose sum equals the target.

```python
arr = [1,2,3,7,5]
target = 12
```

```python


def subarry_sum(arr, target):
    n = len(arr)
    for i in range(n):
        current_sum = 0
        for j in range(i,n):
            current_sum = current_sum + arr[j]
            if current_sum == target:
                return arr[i:j+1]
       
    return []        



arr = [1,2,3,7,5]
target = 12
subarry_sum(arr, target)
        
```

# Sliding Windlow 


```python
def Subarray_sum(arr,target):
    left =0 
    current_sum = 0 

    for right in range(len(arr)):
        current_sum = current_sum + arr[right]

        while current_sum > target:
            current_sum = current_sum - arr[left]
            left = left + 1 
        if current_sum == target:
            return arr[left:right+1]

    return []

    
arr = [1,2,3,7,5]
target = 12
Subarray_sum(arr,target)
```
