```python
def non_repeating(s):

    freq = {}
    #Counts how many times each character appears.
    for ch in s:
        freq[ch] = freq.get(ch,0)+1
    
    # Checks characters in original order.
    for ch in s:
        if freq[ch] == 1:
            return ch
        
    return None 
s = "leetcode"
non_repeating(s)
```
