# Remove Duplicate Characters from String


```python
s = "programming"

n = list(s)
unique = set(n)
unique
```

```python
def remove_duplicate(s):
    unique = ""
    for ch in s:
        if ch not in unique:
            unique = unique + ch
    return unique 

s = "programming"
remove_duplicate(s)
```

```python
def remove_duplicate(s):
    seen = set()
    result = ""
    for ch in s:
        if ch not in seen:
            result = result + ch 
            seen.add(ch)
    return result

s = "programming"
remove_duplicate(s)   

```
