```python
s = "programming"
a = list(s)
def max_freq(s):
    freq = {}

    for ch in a:
        freq[ch] = freq.get(ch,0) +1

    max_char = None
    max_count = 0
    for ch in freq:
        if freq[ch] > max_count:
            max_count = freq[ch]
            max_char = ch
    return (max_char,max_count)

max_freq(s)

        
```
