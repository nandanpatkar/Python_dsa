```python
s = "abc$%12#@"

def remove_special(s):
    final = ""

    for ch in s:
        if(('a' <= ch <='z')or('A' <= ch <='Z') or ('0' <= ch <='9')):
            final = final+ch
    return final

remove_special(s)
```

```python
s = "abc$%12#@"

def remove_special(s):
    final = ""

    for ch in s:
        if ch.isalnum() :
            final = final+ch
    return final

remove_special(s)
```
