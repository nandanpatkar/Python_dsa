# rotate left

```python

def leftRotate(arr,k):
    new_arr = []
    for i in range(k ,len(arr)):
        new_arr.append(arr[i])

    for j in range(k):
        new_arr.append(arr[j])

    return new_arr

arr = [1,2,3,4,5,6]
k = 2
leftRotate(arr,k)
```

# rotate right 



```python
def rightRotate(arr, k):
    n = len(arr)
    k = k % n
    print(k)

    return arr[n-k:] + arr[:n-k] # arr[start : end] and arr[start : end : step]


arr = [1,2,3,4,5]
k = 2

print(rightRotate(arr, k))
```
