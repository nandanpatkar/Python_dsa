```python


def mergeArrays(arr1, arr2):

    arr = arr1 + arr2
    arr.sort()

    return arr


arr1 = [1,3,5]
arr2 = [2,4,6]

print(mergeArrays(arr1, arr2))


```

```python
def merge_two_sorted_arr(arr1,arr2):

    left = 0 
    right = 0

    merge = []

    while left < len(arr1) and right < len(arr2):
        if arr1[left]<= arr2[right]:
            merge.append(arr1[left])
            left = left +1
        else :
            merge.append(arr2[right])
            right = right + 1
    while left < len(arr1):
            merge.append(arr1[left])
            left = left +1
    while right < len(arr2):   
            merge.append(arr2[right])
            right = right +1

    return merge
```

```python
arr1 = [1,3,5]
arr2 = [2,4,6]
merge_two_sorted_arr(arr1,arr2)
```
