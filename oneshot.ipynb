#  Concept 6b: Scope — Local vs Global Variables


Real-world analogy:
Think of scope like rooms in a house. A variable created inside a room (function) stays in that room — nobody outside can see it. But a variable in the hallway (global) can be seen from every room.

```python
# HALLWAY (global)
# ┌─────────────────────────────────┐
# │  name = "Alice"                 │
# │                                 │
# │  ┌──────────────────────────┐   │
# │  │  ROOM (function)         │   │
# │  │  age = 25  ← local only  │   │
# │  │  can see name ✅          │   │
# │  └──────────────────────────┘   │
# │                                 │
# │  print(age) ❌ ERROR!           │
# └─────────────────────────────────┘
```

```python
# Global variable — lives outside all functions
name = "Alice"

def greet():
    # Local variable — only exists INSIDE this function
    ages = 25
    print(name)   # ✅ can read global variable
    print(ages)    # ✅ works fine inside the function

greet()

```

```python
print(name)       # ✅ works — name is global
print(ages)        # ❌ ERROR! age doesn't exist out here
```

```python
# Modifying a global variable inside a function
count = 0

def increment():
    global count       # tells Python: use the HALLWAY variable
    count = count + 1

increment()
# increment()
print(count)       # 1
```

```python
# Modifying a global variable inside a function
count = 0

def increment():
    global count       # tells Python: use the HALLWAY variable
    count = count + 1

increment()
increment()
print(count)       # 2
```

```python
# What to remember:

# Variables inside a function are local — they disappear when the function ends
# Functions can read globals freely, but to change one, use global keyword
# Avoid overusing global — it makes code messy. Prefer return instead!
```

```python
x = 10

def change():
    x = 99
    print(x)

change()
print(x)
```

# *args and **kwargs — Flexible Inputs

Real-world analogy:
Think of *args like a party invitation — you don't know how many people are coming, so you just say "everyone's welcome!" And **kwargs is like a hotel check-in form — guests fill in their own named fields (name, room type, floor).

```python
# *args — any number of positional inputs
# ┌─────────────────────────────────────┐
# │  add(1, 2, 3, 4, 5)                 │
# │       ↓  ↓  ↓  ↓  ↓                │
# │  args = (1, 2, 3, 4, 5)  ← tuple   │
# └─────────────────────────────────────┘

# **kwargs — any number of NAMED inputs
# ┌─────────────────────────────────────┐
# │  info(name="Alice", age=25)         │
# │         ↓              ↓           │
# │  kwargs = {"name":"Alice","age":25} │
# └─────────────────────────────────────┘
```

```python
# *args — accepts ANY number of positional arguments
def add_all(*args):
    print(args)          # (1, 2, 3, 4) ← it's a tuple!
    total = 0
    for num in args:     # loop through all inputs
        total += num     # same as total = total + num
    return total

print(add_all(1, 2))           # 3
print(add_all(1, 2, 3, 4, 5)) # 15  ← works with any count!


# **kwargs — accepts ANY number of NAMED arguments
def show_info(**kwargs):
    print(kwargs)         # {'name': 'Alice', 'age': 25}
    for key, value in kwargs.items():
        print(key, "→", value)

show_info(name="Alice", age=25)
# name → Alice
# age  → 25


# Combining both — order matters: normal, *args, **kwargs
def mixed(greeting, *args, **kwargs):
    print(greeting)        # Hello!
    print(args)            # (1, 2, 3)
    print(kwargs)          # {'city': 'Paris'}

mixed("Hello!", 1, 2, 3, city="Paris")
```

# Lambda Functions — One-Liners

Real-world analogy:
A lambda is like a post-it note function — too small to deserve a full name and definition, just a quick throwawa

```python
# Regular function:               Lambda equivalent:
# ┌───────────────────┐          ┌──────────────────────────┐
# │ def square(x):    │    →→→   │ square = lambda x: x * x │
# │     return x * x  │          └──────────────────────────┘
# └───────────────────┘               same thing, one line!
```

```python
# Regular function
def square(x):
    return x * x

# Exact same thing as a lambda
square = lambda x: x * x

print(square(5))    # 25


# Lambda with two inputs
add = lambda a, b: a + b
print(add(3, 4))    # 7


# Where lambdas SHINE — inside other functions!
numbers = [5, 2, 8, 1, 9, 3]

# sorted() with a custom key using lambda
sorted_nums = sorted(numbers)
print(sorted_nums)             # [1, 2, 3, 5, 8, 9]

# Sort a list of tuples by the SECOND item
pairs = [(1, 9), (3, 2), (5, 5)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)            # [(3, 2), (5, 5), (1, 9)]


# filter() — keep only items where lambda returns True
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)                   # [2, 4, 6]


# map() — apply a function to every item
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)                 # [2, 4, 6, 8, 10, 12]
```

# String Methods — split, join, upper, lower, replace

```python
# sentence = "hello world"

#   upper()          lower()           split()
# ┌───────────┐    ┌───────────┐    ┌─────────────────┐
# │ "HELLO    │    │ "hello    │    │ ["hello","world"]│
# │  WORLD"   │    │  world"   │    │  ← list!        │
# └───────────┘    └───────────┘    └─────────────────┘

#   replace()                join()
# ┌──────────────────┐    ┌──────────────────────┐
# │ "hello python"   │    │ "hello-world"        │
# │ "world"→"python" │    │ "-".join([...])      │
# └──────────────────┘    └──────────────────────┘
```

# Dictionaries — key-value, get, update

Real-world analogy:
A dictionary is like a real dictionary book — you look up a word (key) and get its definition (value). Instead of searching page by page, you jump directly to what you need. Super fast! 

<!-- student = {"name": "Alice", "age": 25, "grade": "A"}

      KEY       VALUE
   ┌────────┬───────────┐
   │ "name" │  "Alice"  │
   ├────────┼───────────┤
   │ "age"  │    25     │
   ├────────┼───────────┤
   │"grade" │    "A"    │
   └────────┴───────────┘
   
   student["name"] → "Alice"  ⚡ instant lookup! -->

```python
# Creating a dictionary — key: value pairs inside {}
student = {"name": "Alice", "age": 25, "grade": "A"}

# Accessing values by key
print(student["name"])         # Alice
print(student["age"])          # 25

# ❌ Accessing a missing key crashes!
# print(student["phone"])      # KeyError!

# ✅ get() — safe lookup, returns None if key missing
print(student.get("phone"))    # None  ← no crash!
print(student.get("phone", "N/A"))  # N/A  ← custom default

# Adding a new key-value pair
student["city"] = "Delhi"
print(student)   # {"name":"Alice","age":25,"grade":"A","city":"Delhi"}

# Updating an existing value
student["grade"] = "A+"
print(student["grade"])        # A+

# Looping through a dictionary
for key, value in student.items():
    print(key, "→", value)
# name → Alice
# age  → 25
# grade → A+
# city → Delhi

# Check if a key exists
print("name" in student)       # True
print("phone" in student)      # False

# Common dictionary methods
print(student.keys())          # dict_keys(['name', 'age', 'grade', 'city'])
print(student.values())        # dict_values(['Alice', 25, 'A+', 'Delhi'])
```

# Sets — Uniqueness, Union, Intersection

A set is like a guest list at an exclusive party — each person can only appear once, no duplicates allowed! And you can easily compare two guest lists to find who's on both, or combine them into one big list.

```python
# list  = [1, 2, 2, 3, 3, 3]   ← allows duplicates
# set   = {1, 2, 3}             ← duplicates removed automatically!

#      SET A          SET B
#    ┌────────┐     ┌────────┐
#    │ 1  2   │     │  2  3  │
#    │    ↑───┼─────┼───↑    │
#    │  3     │     │     4  │
#    └────────┘     └────────┘
#    union = {1,2,3,4}   intersection = {2,3}
```

```python
# Creating a set — curly braces like dict, but NO key:value pairs
fruits = {"apple", "banana", "cherry"}
print(fruits)              # {'apple', 'banana', 'cherry'}

# Duplicates are automatically removed!
numbers = {1, 2, 2, 3, 3, 3}
print(numbers)             # {1, 2, 3}

# Convert a list to a set to remove duplicates
my_list = [1, 2, 2, 3, 3, 3]
my_set = set(my_list)
print(my_set)              # {1, 2, 3}

# Add and remove items
fruits.add("mango")        # adds one item
fruits.remove("banana")    # removes item ❌ crashes if missing!
fruits.discard("banana")   # ✅ safer — no crash if missing

# Check membership — sets are SUPER fast at this!
print("apple" in fruits)   # True
print("grape" in fruits)   # False

# Union — all items from BOTH sets (no duplicates)
a = {1, 2, 3}
b = {2, 3, 4}
print(a | b)               # {1, 2, 3, 4}
print(a.union(b))          # same thing

# Intersection — only items in BOTH sets
print(a & b)               # {2, 3}
print(a.intersection(b))   # same thing

# Difference — items in a but NOT in b
print(a - b)               # {1}
print(a.difference(b))     # same thing
```

# Tuples — Immutability & Unpacking

A tuple is like a printed certificate — once it's printed, you cannot change it. A list is like a whiteboard (erasable), but a tuple is permanent ink. Great for data that should never be modified, like coordinates, RGB colors, or dates!

```python
# list  = [1, 2, 3]   ← can change 
# tuple = (1, 2, 3)   ← frozen, cannot change 

#       UNPACKING a tuple:
#       point = (10, 20)
#          ↓      ↓
#          x      y
#       x = 10, y = 20  ← each value goes into its own variable
```

```python
# Creating a tuple — round brackets
point = (10, 20)
colors = ("red", "green", "blue")
single = (42,)             # ← comma needed for single item tuple!

# Accessing — same as lists, use index
print(point[0])            # 10
print(colors[-1])          # blue

# ❌ Tuples are IMMUTABLE — cannot change them!
# point[0] = 99              # TypeError: 'tuple' object does not support item assignment

# Unpacking — assign each value to its own variable
x, y = point
print(x)                   # 10
print(y)                   # 20

# Unpacking works with any iterable!
r, g, b = colors
print(r)                   # red

# Swap two variables elegantly using tuple unpacking
a = 5
b = 10
a, b = b, a                # ✅ no temp variable needed!
print(a, b)                # 10 5

# Tuple of tuples — common in LeetCode!
students = [("Alice", 90), ("Bob", 85), ("Charlie", 92)]
for name, score in students:   # unpack directly in loop!
    print(name, "→", score)
# Alice → 90
# Bob → 85
# Charlie → 92
```

```python
def second_largest(arr):
    largest = float('-inf')
    second = float('-inf')

    for num in arr:
        if num > largest:
            second = largest
            largest = num

        elif num > second and num != largest:
            second = num

    return second


arr = [3, 2, 1, 5, 6, 4]
print(second_largest(arr))
```

```python
import heapq

def kth_largest(arr, k):
    heap = []

    for num in arr:
        heapq.heappush(heap, num)

        if len(heap) > k:
            heapq.heappop(heap)

    return heap[0]


arr = [3,2,1,5,6,4]
k = 2

print(kth_largest(arr,k))
```
