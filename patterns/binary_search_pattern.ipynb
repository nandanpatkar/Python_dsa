# 6️⃣ Binary Search PatternEach question has both **Brute Force** and **Optimal** approach.

## Binary Search

```python
# Brute Force Approachdef binary_search_bruteforce(nums, target):    for i, x in enumerate(nums):        if x == target:            return i    return -1print(binary_search_bruteforce([-1,0,3,5,9,12], 9))
```

```python
# Optimal Approachdef binary_search(nums, target):    l, r = 0, len(nums)-1    while l <= r:        m = (l + r) // 2        if nums[m] == target: return m        if nums[m] < target: l = m + 1        else: r = m - 1    return -1print(binary_search([-1,0,3,5,9,12], 9))
```

## Search in Rotated Sorted Array

```python
# Brute Force Approachdef search_rotated_bruteforce(nums, target):    for i, x in enumerate(nums):        if x == target:            return i    return -1print(search_rotated_bruteforce([4,5,6,7,0,1,2], 0))
```

```python
# Optimal Approachdef search_rotated(nums, target):    l, r = 0, len(nums)-1    while l <= r:        m = (l + r) // 2        if nums[m] == target: return m        if nums[l] <= nums[m]:            if nums[l] <= target < nums[m]: r = m - 1            else: l = m + 1        else:            if nums[m] < target <= nums[r]: l = m + 1            else: r = m - 1    return -1print(search_rotated([4,5,6,7,0,1,2], 0))
```

## Find Peak Element

```python
# Brute Force Approachdef find_peak_bruteforce(nums):    for i in range(len(nums)):        left = nums[i-1] if i > 0 else float('-inf')        right = nums[i+1] if i < len(nums)-1 else float('-inf')        if nums[i] > left and nums[i] > right:            return iprint(find_peak_bruteforce([1,2,1,3,5,6,4]))
```

```python
# Optimal Approachdef find_peak(nums):    l, r = 0, len(nums)-1    while l < r:        m = (l + r) // 2        if nums[m] < nums[m+1]: l = m + 1        else: r = m    return lprint(find_peak([1,2,1,3,5,6,4]))
```

## Median of Two Sorted Arrays

```python
# Brute Force Approachdef find_median_sorted_arrays_bruteforce(a, b):    arr = sorted(a + b)    n = len(arr)    if n % 2:        return arr[n // 2]    return (arr[n // 2 - 1] + arr[n // 2]) / 2print(find_median_sorted_arrays_bruteforce([1,3], [2]))
```

```python
# Optimal Approachdef find_median_sorted_arrays(a, b):    if len(a) > len(b):        a, b = b, a    total = len(a) + len(b)    half = total // 2    l, r = 0, len(a)    while True:        i = (l + r) // 2        j = half - i        aL = a[i-1] if i > 0 else float('-inf')        aR = a[i] if i < len(a) else float('inf')        bL = b[j-1] if j > 0 else float('-inf')        bR = b[j] if j < len(b) else float('inf')        if aL <= bR and bL <= aR:            if total % 2: return min(aR, bR)            return (max(aL, bL) + min(aR, bR)) / 2        elif aL > bR: r = i - 1        else: l = i + 1print(find_median_sorted_arrays([1,3], [2]))
```
