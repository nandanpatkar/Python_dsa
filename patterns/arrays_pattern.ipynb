Each question has both **Brute Force** and **Optimal** approach.

## Two Sum

```python
# Brute Force Approach
def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

print(two_sum_bruteforce([2,7,11,15], 9))
```

```python
# Optimal Approach
def two_sum(nums, target):
    seen = {}
    for i, x in enumerate(nums):
        if target - x in seen:
            return [seen[target - x], i]
        seen[x] = i
    return []

print(two_sum([2,7,11,15], 9))
```

## Best Time to Buy and Sell Stock

```python
# Brute Force Approach
def max_profit_bruteforce(prices):
    best = 0
    for i in range(len(prices)):
        for j in range(i+1, len(prices)):
            best = max(best, prices[j] - prices[i])
    return best

print(max_profit_bruteforce([7,1,5,3,6,4]))
```

```python
# Optimal Approach
def max_profit(prices):
    min_price = float('inf')
    best = 0
    for p in prices:
        min_price = min(min_price, p)
        best = max(best, p - min_price)
    return best

print(max_profit([7,1,5,3,6,4]))
```

## Maximum Subarray (Kadane’s Algorithm)

```python
# Brute Force Approach
def max_subarray_bruteforce(nums):
    best = nums[0]
    for i in range(len(nums)):
        total = 0
        for j in range(i, len(nums)):
            total += nums[j]
            best = max(best, total)
    return best

print(max_subarray_bruteforce([-2,1,-3,4,-1,2,1,-5,4]))
```

```python
# Optimal Approach
def max_subarray(nums):
    curr = best = nums[0]
    for x in nums[1:]:
        curr = max(x, curr + x)
        best = max(best, curr)
    return best

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

## Product of Array Except Itself

```python
# Brute Force Approach
def product_except_self_bruteforce(nums):
    out = []
    for i in range(len(nums)):
        p = 1
        for j in range(len(nums)):
            if i != j:
                p *= nums[j]
        out.append(p)
    return out

print(product_except_self_bruteforce([1,2,3,4]))
```

```python
# Optimal Approach
def product_except_self(nums):
    n = len(nums)
    out = [1] * n
    pref = 1
    for i in range(n):
        out[i] = pref
        pref *= nums[i]
    suff = 1
    for i in range(n-1, -1, -1):
        out[i] *= suff
        suff *= nums[i]
    return out

print(product_except_self([1,2,3,4]))
```

## Move Zeroes

```python
# Brute Force Approach
def move_zeroes_bruteforce(nums):
    nums = nums[:]
    i = 0
    while i < len(nums):
        if nums[i] == 0:
            nums.pop(i)
            nums.append(0)
        else:
            i += 1
    return nums

print(move_zeroes_bruteforce([0,1,0,3,12]))
```

```python
# Optimal Approach
def move_zeroes(nums):
    left = 0
    for right in range(len(nums)):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
    return nums

print(move_zeroes([0,1,0,3,12]))
```

## Missing Number

```python
# Brute Force Approach
def missing_number_bruteforce(nums):
    n = len(nums)
    for x in range(n+1):
        if x not in nums:
            return x

print(missing_number_bruteforce([3,0,1]))
```

```python
# Optimal Approach
def missing_number(nums):
    n = len(nums)
    return n * (n + 1) // 2 - sum(nums)

print(missing_number([3,0,1]))
```

```python
def missing_number_gauss(nums):
    n = len(nums)

    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)

    return expected_sum - actual_sum


print(missing_number_gauss([3,0,1]))
```

## Find Duplicate Number

```python
# Brute Force Approach
def find_duplicate_bruteforce(nums):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] == nums[j]:
                return nums[i]
    return -1

print(find_duplicate_bruteforce([1,3,4,2,2]))
```

```python
# Optimal Approach
def find_duplicate(nums):
    slow = fast = nums[0]
    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break
    p = nums[0]
    while p != slow:
        p = nums[p]
        slow = nums[slow]
    return slow

print(find_duplicate([1,3,4,2,2]))
```

## Rotate Array

```python
# Brute Force Approach
def rotate_bruteforce(nums, k):
    nums = nums[:]
    n = len(nums)
    k %= n
    for _ in range(k):
        nums.insert(0, nums.pop())
    return nums

print(rotate_bruteforce([1,2,3,4,5,6,7], 3))
```

```python
# Optimal Approach
def rotate(nums, k):
    k %= len(nums)
    nums[:] = nums[-k:] + nums[:-k]
    return nums

print(rotate([1,2,3,4,5,6,7], 3))
```

## Merge Sorted Arrays

```python
# Brute Force Approach
def merge_sorted_bruteforce(a, b):
    arr = a + b
    arr.sort()
    return arr

print(merge_sorted_bruteforce([1,2,4], [1,3,4]))
```

```python
# Optimal Approach
def merge_sorted(a, b):
    i = j = 0
    out = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            out.append(a[i]); i += 1
        else:
            out.append(b[j]); j += 1
    out.extend(a[i:]); out.extend(b[j:])
    return out

print(merge_sorted([1,2,4], [1,3,4]))
```

## Longest Consecutive Sequence

```python
# Brute Force Approach
def longest_consecutive_bruteforce(nums):
    if not nums:
        return 0
    nums = sorted(set(nums))
    best = curr = 1
    for i in range(1, len(nums)):
        if nums[i] == nums[i-1] + 1:
            curr += 1
            best = max(best, curr)
        else:
            curr = 1
    return best

print(longest_consecutive_bruteforce([100,4,200,1,3,2]))
```

```python
# Optimal Approach
def longest_consecutive(nums):
    s = set(nums)
    best = 0
    for x in s:
        if x - 1 not in s:
            y = x
            while y in s:
                y += 1
            best = max(best, y - x)
    return best

print(longest_consecutive([100,4,200,1,3,2]))
```
