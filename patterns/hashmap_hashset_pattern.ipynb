## Two Sum

```python
# Brute Force Approachdef two_sum_bruteforce(nums, target):    for i in range(len(nums)):        for j in range(i+1, len(nums)):            if nums[i] + nums[j] == target:                return [i, j]    return []print(two_sum_bruteforce([2,7,11,15], 9))
```

```python
# Optimal Approachdef two_sum(nums, target):    seen = {}    for i, x in enumerate(nums):        if target - x in seen:            return [seen[target - x], i]        seen[x] = i    return []print(two_sum([2,7,11,15], 9))
```

## Valid Anagram

```python
# Brute Force Approachdef is_anagram_bruteforce(s, t):    return sorted(s) == sorted(t)print(is_anagram_bruteforce('anagram', 'nagaram'))
```

```python
# Optimal Approachdef is_anagram(s, t):    if len(s) != len(t):        return False    freq = {}    for ch in s:        freq[ch] = freq.get(ch, 0) + 1    for ch in t:        if ch not in freq:            return False        freq[ch] -= 1        if freq[ch] == 0:            del freq[ch]    return len(freq) == 0print(is_anagram('anagram', 'nagaram'))
```

## Group Anagrams

```python
# Brute Force Approachdef group_anagrams_bruteforce(strs):    groups = []    used = [False] * len(strs)    for i in range(len(strs)):        if used[i]:            continue        g = [strs[i]]        used[i] = True        for j in range(i+1, len(strs)):            if not used[j] and sorted(strs[i]) == sorted(strs[j]):                g.append(strs[j])                used[j] = True        groups.append(g)    return groupsprint(group_anagrams_bruteforce(['eat','tea','tan','ate','nat','bat']))
```

```python
# Optimal Approachfrom collections import defaultdictdef group_anagrams(strs):    d = defaultdict(list)    for s in strs:        d[tuple(sorted(s))].append(s)    return list(d.values())print(group_anagrams(['eat','tea','tan','ate','nat','bat']))
```

## First Unique Character in String

```python
# Brute Force Approachdef first_uniq_char_bruteforce(s):    for i in range(len(s)):        if s.count(s[i]) == 1:            return i    return -1print(first_uniq_char_bruteforce('leetcode'))
```

```python
# Optimal Approachdef first_uniq_char(s):    freq = {}    for ch in s:        freq[ch] = freq.get(ch, 0) + 1    for i, ch in enumerate(s):        if freq[ch] == 1:            return i    return -1print(first_uniq_char('leetcode'))
```

## Longest Consecutive Sequence

```python
# Brute Force Approachdef longest_consecutive_bruteforce(nums):    if not nums:        return 0    nums = sorted(set(nums))    best = curr = 1    for i in range(1, len(nums)):        if nums[i] == nums[i-1] + 1:            curr += 1            best = max(best, curr)        else:            curr = 1    return bestprint(longest_consecutive_bruteforce([100,4,200,1,3,2]))
```

```python
# Optimal Approachdef longest_consecutive(nums):    s = set(nums)    best = 0    for x in s:        if x - 1 not in s:            y = x            while y in s:                y += 1            best = max(best, y - x)    return bestprint(longest_consecutive([100,4,200,1,3,2]))
```

## Top K Frequent Elements

```python
# Brute Force Approachdef top_k_frequent_bruteforce(nums, k):    uniq = list(set(nums))    uniq.sort(key=lambda x: nums.count(x), reverse=True)    return uniq[:k]print(top_k_frequent_bruteforce([1,1,1,2,2,3], 2))
```

```python
# Optimal Approachimport heapqdef top_k_frequent(nums, k):    freq = {}    for x in nums:        freq[x] = freq.get(x, 0) + 1    return [x for x, _ in heapq.nlargest(k, freq.items(), key=lambda p: p[1])]print(top_k_frequent([1,1,1,2,2,3], 2))
```
