# 2️⃣ Two Pointer PatternEach question has both **Brute Force** and **Optimal** approach.

## Valid Palindrome

```python
# Brute Force Approachdef is_palindrome_bruteforce(s):    cleaned = ''.join(ch.lower() for ch in s if ch.isalnum())    return cleaned == cleaned[::-1]print(is_palindrome_bruteforce('A man, a plan, a canal: Panama'))
```

```python
# Optimal Approachdef is_palindrome(s):    l, r = 0, len(s) - 1    while l < r:        while l < r and not s[l].isalnum(): l += 1        while l < r and not s[r].isalnum(): r -= 1        if s[l].lower() != s[r].lower(): return False        l += 1; r -= 1    return Trueprint(is_palindrome('A man, a plan, a canal: Panama'))
```

## Two Sum II (Sorted array)

```python
# Brute Force Approachdef two_sum_sorted_bruteforce(nums, target):    for i in range(len(nums)):        for j in range(i+1, len(nums)):            if nums[i] + nums[j] == target:                return [i+1, j+1]    return []print(two_sum_sorted_bruteforce([2,7,11,15], 9))
```

```python
# Optimal Approachdef two_sum_sorted(nums, target):    l, r = 0, len(nums)-1    while l < r:        s = nums[l] + nums[r]        if s == target: return [l+1, r+1]        if s < target: l += 1        else: r -= 1    return []print(two_sum_sorted([2,7,11,15], 9))
```

## Container With Most Water

```python
# Brute Force Approachdef max_area_bruteforce(height):    best = 0    for i in range(len(height)):        for j in range(i+1, len(height)):            best = max(best, min(height[i], height[j]) * (j - i))    return bestprint(max_area_bruteforce([1,8,6,2,5,4,8,3,7]))
```

```python
# Optimal Approachdef max_area(height):    l, r = 0, len(height)-1    best = 0    while l < r:        best = max(best, min(height[l], height[r]) * (r - l))        if height[l] < height[r]: l += 1        else: r -= 1    return bestprint(max_area([1,8,6,2,5,4,8,3,7]))
```

## Remove Duplicates from Sorted Array

```python
# Brute Force Approachdef remove_duplicates_bruteforce(nums):    uniq = []    for x in nums:        if x not in uniq:            uniq.append(x)    nums[:len(uniq)] = uniq    return len(uniq)arr = [0,0,1,1,1,2,2,3,3,4]k = remove_duplicates_bruteforce(arr)print(k, arr[:k])
```

```python
# Optimal Approachdef remove_duplicates(nums):    if not nums: return 0    w = 1    for r in range(1, len(nums)):        if nums[r] != nums[r-1]:            nums[w] = nums[r]            w += 1    return warr = [0,0,1,1,1,2,2,3,3,4]k = remove_duplicates(arr)print(k, arr[:k])
```

## 3Sum

```python
# Brute Force Approachdef three_sum_bruteforce(nums):    n = len(nums)    ans = set()    for i in range(n):        for j in range(i+1, n):            for k in range(j+1, n):                if nums[i] + nums[j] + nums[k] == 0:                    ans.add(tuple(sorted((nums[i], nums[j], nums[k]))))    return [list(t) for t in ans]print(three_sum_bruteforce([-1,0,1,2,-1,-4]))
```

```python
# Optimal Approachdef three_sum(nums):    nums.sort()    out = []    for i in range(len(nums)-2):        if i > 0 and nums[i] == nums[i-1]:            continue        l, r = i + 1, len(nums) - 1        while l < r:            s = nums[i] + nums[l] + nums[r]            if s == 0:                out.append([nums[i], nums[l], nums[r]])                l += 1; r -= 1                while l < r and nums[l] == nums[l-1]: l += 1                while l < r and nums[r] == nums[r+1]: r -= 1            elif s < 0: l += 1            else: r -= 1    return outprint(three_sum([-1,0,1,2,-1,-4]))
```

## Trapping Rain Water

```python
# Brute Force Approachdef trap_bruteforce(height):    water = 0    for i in range(len(height)):        left_max = max(height[:i+1])        right_max = max(height[i:])        water += min(left_max, right_max) - height[i]    return waterprint(trap_bruteforce([0,1,0,2,1,0,1,3,2,1,2,1]))
```

```python
# Optimal Approachdef trap(height):    l, r = 0, len(height)-1    lmax = rmax = 0    water = 0    while l < r:        if height[l] < height[r]:            lmax = max(lmax, height[l])            water += lmax - height[l]            l += 1        else:            rmax = max(rmax, height[r])            water += rmax - height[r]            r -= 1    return waterprint(trap([0,1,0,2,1,0,1,3,2,1,2,1]))
```

## Squares of Sorted Array

```python
# Brute Force Approachdef sorted_squares_bruteforce(nums):    out = [x*x for x in nums]    out.sort()    return outprint(sorted_squares_bruteforce([-4,-1,0,3,10]))
```

```python
# Optimal Approachdef sorted_squares(nums):    l, r = 0, len(nums)-1    out = [0] * len(nums)    i = len(nums)-1    while l <= r:        if abs(nums[l]) > abs(nums[r]):            out[i] = nums[l] * nums[l]            l += 1        else:            out[i] = nums[r] * nums[r]            r -= 1        i -= 1    return outprint(sorted_squares([-4,-1,0,3,10]))
```
