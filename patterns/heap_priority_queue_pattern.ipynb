# Heap / Priority Queue PatternEach question has both **Brute Force** and **Optimal** approach.

## Kth Largest Element in Array

```python
# Brute Force Approachdef kth_largest_bruteforce(nums, k):    arr = sorted(nums, reverse=True)    return arr[k-1]print(kth_largest_bruteforce([3,2,1,5,6,4], 2))
```

```python
# Optimal Approachimport heapqdef kth_largest(nums, k):    heap = []    for x in nums:        heapq.heappush(heap, x)        if len(heap) > k:            heapq.heappop(heap)    return heap[0]print(kth_largest([3,2,1,5,6,4], 2))
```

## Top K Frequent Elements

```python
# Brute Force Approachdef top_k_frequent_bruteforce(nums, k):    uniq = list(set(nums))    uniq.sort(key=lambda x: nums.count(x), reverse=True)    return uniq[:k]print(top_k_frequent_bruteforce([1,1,1,2,2,3], 2))
```

```python
# Optimal Approachimport heapqdef top_k_frequent(nums, k):    freq = {}    for x in nums:        freq[x] = freq.get(x, 0) + 1    return [x for x, _ in heapq.nlargest(k, freq.items(), key=lambda p: p[1])]print(top_k_frequent([1,1,1,2,2,3], 2))
```

## Merge K Sorted Lists

```python
# Brute Force Approachdef merge_k_sorted_lists_bruteforce(lists):    arr = []    for l in lists:        arr.extend(l)    arr.sort()    return arrprint(merge_k_sorted_lists_bruteforce([[1,4,5],[1,3,4],[2,6]]))
```

```python
# Optimal Approachimport heapqdef merge_k_sorted_lists(lists):    heap = []    for i, arr in enumerate(lists):        if arr:            heapq.heappush(heap, (arr[0], i, 0))    out = []    while heap:        val, li, idx = heapq.heappop(heap)        out.append(val)        idx += 1        if idx < len(lists[li]):            heapq.heappush(heap, (lists[li][idx], li, idx))    return outprint(merge_k_sorted_lists([[1,4,5],[1,3,4],[2,6]]))
```

## Find Median from Data Stream

```python
# Brute Force Approachclass MedianFinderBruteforce:    def __init__(self):        self.data = []    def addNum(self, num):        self.data.append(num)    def findMedian(self):        arr = sorted(self.data)        n = len(arr)        if n % 2:            return arr[n // 2]        return (arr[n // 2 - 1] + arr[n // 2]) / 2mf = MedianFinderBruteforce()for x in [1,2,3,4]: mf.addNum(x)print(mf.findMedian())
```

```python
# Optimal Approachimport heapqclass MedianFinder:    def __init__(self):        self.small = []        self.large = []    def addNum(self, num):        heapq.heappush(self.small, -num)        if self.large and -self.small[0] > self.large[0]:            heapq.heappush(self.large, -heapq.heappop(self.small))        if len(self.small) > len(self.large) + 1:            heapq.heappush(self.large, -heapq.heappop(self.small))        if len(self.large) > len(self.small):            heapq.heappush(self.small, -heapq.heappop(self.large))    def findMedian(self):        if len(self.small) > len(self.large):            return -self.small[0]        return (-self.small[0] + self.large[0]) / 2mf = MedianFinder()for x in [1,2,3,4]: mf.addNum(x)print(mf.findMedian())
```

## K Closest Points to Origin

```python
# Brute Force Approachdef k_closest_bruteforce(points, k):    points = sorted(points, key=lambda p: p[0]*p[0] + p[1]*p[1])    return points[:k]print(k_closest_bruteforce([[1,3],[-2,2],[5,8]], 2))
```

```python
# Optimal Approachimport heapqdef k_closest(points, k):    return heapq.nsmallest(k, points, key=lambda p: p[0]*p[0] + p[1]*p[1])print(k_closest([[1,3],[-2,2],[5,8]], 2))
```
