# 3️⃣ Sliding Window PatternEach question has both **Brute Force** and **Optimal** approach.

## Maximum Sum Subarray of Size K

```python
# Brute Force Approachdef max_sum_k_bruteforce(nums, k):    best = float('-inf')    for i in range(len(nums)-k+1):        best = max(best, sum(nums[i:i+k]))    return bestprint(max_sum_k_bruteforce([2,1,5,1,3,2], 3))
```

```python
# Optimal Approachdef max_sum_k(nums, k):    window = sum(nums[:k])    best = window    for i in range(k, len(nums)):        window += nums[i] - nums[i-k]        best = max(best, window)    return bestprint(max_sum_k([2,1,5,1,3,2], 3))
```

## Longest Substring Without Repeating Characters

```python
# Brute Force Approachdef length_of_longest_substring_bruteforce(s):    best = 0    for i in range(len(s)):        seen = set()        for j in range(i, len(s)):            if s[j] in seen:                break            seen.add(s[j])            best = max(best, j - i + 1)    return bestprint(length_of_longest_substring_bruteforce('abcabcbb'))
```

```python
# Optimal Approachdef length_of_longest_substring(s):    seen = {}    l = 0    best = 0    for r, ch in enumerate(s):        if ch in seen and seen[ch] >= l:            l = seen[ch] + 1        seen[ch] = r        best = max(best, r - l + 1)    return bestprint(length_of_longest_substring('abcabcbb'))
```

## Longest Repeating Character Replacement

```python
# Brute Force Approachdef character_replacement_bruteforce(s, k):    best = 0    for i in range(len(s)):        freq = {}        maxf = 0        for j in range(i, len(s)):            freq[s[j]] = freq.get(s[j], 0) + 1            maxf = max(maxf, freq[s[j]])            if (j - i + 1) - maxf <= k:                best = max(best, j - i + 1)    return bestprint(character_replacement_bruteforce('AABABBA', 1))
```

```python
# Optimal Approachdef character_replacement(s, k):    count = {}    l = 0    maxf = 0    best = 0    for r, ch in enumerate(s):        count[ch] = count.get(ch, 0) + 1        maxf = max(maxf, count[ch])        while (r - l + 1) - maxf > k:            count[s[l]] -= 1            l += 1        best = max(best, r - l + 1)    return bestprint(character_replacement('AABABBA', 1))
```

```python
# Optimal Approachdef min_subarray_len(target, nums):    l = 0    total = 0    ans = float('inf')    for r, x in enumerate(nums):        total += x        while total >= target:            ans = min(ans, r - l + 1)            total -= nums[l]            l += 1    return 0 if ans == float('inf') else ansprint(min_subarray_len(7, [2,3,1,2,4,3]))
```

## Maximum Number of Vowels in Substring

```python
# Brute Force Approachdef max_vowels_bruteforce(s, k):    vowels = set('aeiou')    best = 0    for i in range(len(s)-k+1):        cnt = 0        for ch in s[i:i+k]:            if ch in vowels:                cnt += 1        best = max(best, cnt)    return bestprint(max_vowels_bruteforce('abciiidef', 3))
```

```python
# Optimal Approachdef max_vowels(s, k):    vowels = set('aeiou')    curr = sum(1 for ch in s[:k] if ch in vowels)    best = curr    for i in range(k, len(s)):        curr += (s[i] in vowels) - (s[i-k] in vowels)        best = max(best, curr)    return bestprint(max_vowels('abciiidef', 3))
```

## Max Consecutive Ones

```python
# Brute Force Approachdef find_max_consecutive_ones_bruteforce(nums):    best = 0    for i in range(len(nums)):        if nums[i] == 1:            j = i            while j < len(nums) and nums[j] == 1:                j += 1            best = max(best, j - i)    return bestprint(find_max_consecutive_ones_bruteforce([1,1,0,1,1,1]))
```

## Permutation in String

```python
# Brute Force Approachfrom collections import Counterdef check_inclusion_bruteforce(s1, s2):    need = Counter(s1)    k = len(s1)    for i in range(len(s2)-k+1):        if Counter(s2[i:i+k]) == need:            return True    return Falseprint(check_inclusion_bruteforce('ab', 'eidbaooo'))
```

```python
# Optimal Approachfrom collections import Counterdef check_inclusion(s1, s2):    need = Counter(s1)    win = Counter()    l = 0    for r, ch in enumerate(s2):        win[ch] += 1        if r - l + 1 > len(s1):            left = s2[l]            win[left] -= 1            if win[left] == 0:                del win[left]            l += 1        if win == need:            return True    return Falseprint(check_inclusion('ab', 'eidbaooo'))
```
