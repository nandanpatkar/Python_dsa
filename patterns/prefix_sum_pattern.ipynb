## Range Sum Query

```python
# Brute Force Approachclass NumArrayBruteforce:    def __init__(self, nums):        self.nums = nums    def sumRange(self, left, right):        s = 0        for i in range(left, right + 1):            s += self.nums[i]        return sna = NumArrayBruteforce([-2,0,3,-5,2,-1])print(na.sumRange(0,2), na.sumRange(2,5))
```

```python
# Optimal Approachclass NumArray:    def __init__(self, nums):        self.pref = [0]        for x in nums:            self.pref.append(self.pref[-1] + x)    def sumRange(self, left, right):        return self.pref[right + 1] - self.pref[left]na = NumArray([-2,0,3,-5,2,-1])print(na.sumRange(0,2), na.sumRange(2,5))
```

## Subarray Sum Equals K

```python
# Brute Force Approachdef subarray_sum_bruteforce(nums, k):    count = 0    for i in range(len(nums)):        s = 0        for j in range(i, len(nums)):            s += nums[j]            if s == k:                count += 1    return countprint(subarray_sum_bruteforce([1,1,1], 2))
```

```python
# Optimal Approachdef subarray_sum(nums, k):    count = 0    pref = 0    seen = {0:1}    for x in nums:        pref += x        count += seen.get(pref - k, 0)        seen[pref] = seen.get(pref, 0) + 1    return countprint(subarray_sum([1,1,1], 2))
```

## Running Sum of Array

```python
# Brute Force Approachdef running_sum_bruteforce(nums):    out = []    for i in range(len(nums)):        out.append(sum(nums[:i+1]))    return outprint(running_sum_bruteforce([1,2,3,4]))
```

```python
# Optimal Approachdef running_sum(nums):    out = []    total = 0    for x in nums:        total += x        out.append(total)    return outprint(running_sum([1,2,3,4]))
```

## Continuous Subarray Sum

```python
# Brute Force Approachdef check_subarray_sum_bruteforce(nums, k):    for i in range(len(nums)):        s = 0        for j in range(i, len(nums)):            s += nums[j]            if j - i + 1 >= 2 and s % k == 0:                return True    return Falseprint(check_subarray_sum_bruteforce([23,2,4,6,7], 6))
```

```python
# Optimal Approachdef check_subarray_sum(nums, k):    rem_index = {0:-1}    pref = 0    for i, x in enumerate(nums):        pref += x        rem = pref % k        if rem in rem_index:            if i - rem_index[rem] > 1:                return True        else:            rem_index[rem] = i    return Falseprint(check_subarray_sum([23,2,4,6,7], 6))
```

## Find Pivot Index

```python
# Brute Force Approachdef pivot_index_bruteforce(nums):    for i in range(len(nums)):        if sum(nums[:i]) == sum(nums[i+1:]):            return i    return -1print(pivot_index_bruteforce([1,7,3,6,5,6]))
```

```python
# Optimal Approachdef pivot_index(nums):    total = sum(nums)    left = 0    for i, x in enumerate(nums):        if left == total - left - x:            return i        left += x    return -1print(pivot_index([1,7,3,6,5,6]))
```

## Maximum Size Subarray Sum Equals K

```python
# Brute Force Approachdef max_subarray_len_bruteforce(nums, k):    best = 0    for i in range(len(nums)):        s = 0        for j in range(i, len(nums)):            s += nums[j]            if s == k:                best = max(best, j - i + 1)    return bestprint(max_subarray_len_bruteforce([1,-1,5,-2,3], 3))
```

```python
# Optimal Approachdef max_subarray_len(nums, k):    pref = 0    first = {0:-1}    best = 0    for i, x in enumerate(nums):        pref += x        if pref - k in first:            best = max(best, i - first[pref - k])        if pref not in first:            first[pref] = i    return bestprint(max_subarray_len([1,-1,5,-2,3], 3))
```
