# Sets — Uniqueness, Union, Intersection

A set is like a guest list at an exclusive party — each person can only appear once, no duplicates allowed! And you can easily compare two guest lists to find who's on both, or combine them into one big list.

```python
# list  = [1, 2, 2, 3, 3, 3]   ← allows duplicates
# set   = {1, 2, 3}             ← duplicates removed automatically!

#      SET A          SET B
#    ┌────────┐     ┌────────┐
#    │ 1  2   │     │  2  3  │
#    │    ↑───┼─────┼───↑    │
#    │  3     │     │     4  │
#    └────────┘     └────────┘
#    union = {1,2,3,4}   intersection = {2,3}
```

```python
# Creating a set — curly braces like dict, but NO key:value pairs
fruits = {"apple", "banana", "cherry"}
print(fruits)              # {'apple', 'banana', 'cherry'}

# Duplicates are automatically removed!
numbers = {1, 2, 2, 3, 3, 3}
print(numbers)             # {1, 2, 3}

# Convert a list to a set to remove duplicates
my_list = [1, 2, 2, 3, 3, 3]
my_set = set(my_list)
print(my_set)              # {1, 2, 3}

# Add and remove items
fruits.add("mango")        # adds one item
fruits.remove("banana")    # removes item ❌ crashes if missing!
fruits.discard("banana")   # ✅ safer — no crash if missing

# Check membership — sets are SUPER fast at this!
print("apple" in fruits)   # True
print("grape" in fruits)   # False

# Union — all items from BOTH sets (no duplicates)
a = {1, 2, 3}
b = {2, 3, 4}
print(a | b)               # {1, 2, 3, 4}
print(a.union(b))          # same thing

# Intersection — only items in BOTH sets
print(a & b)               # {2, 3}
print(a.intersection(b))   # same thing

# Difference — items in a but NOT in b
print(a - b)               # {1}
print(a.difference(b))     # same thing
```

```python
# What to remember:

# Sets are unordered — you can't index them like my_set[0] ❌
# Sets are blazing fast for in checks — much faster than lists!
# Use sets to remove duplicates from a list in one line: set(my_list)
# | union, & intersection, - difference — think Venn diagrams!
```

```python
class_a = ["Alice", "Bob", "Charlie", "Diana"]
class_b = ["Bob", "Diana", "Eve", "Frank"]
# Using sets, find:

# All unique students across both classes (union)
# Students in both classes (intersection)
#Students only in class A (difference)
```

```python
set_a = set(class_a)
set_b = set(class_b)
```

```python
set_a | set_b
```

```python
set_a & set_b
```

```python
set_a - set_b
```

```python
set_b - set_a
```

```python
# Inbuilt set methods
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print("a:", a)
print("b:", b)
print("union:", a.union(b))
print("intersection:", a.intersection(b))
print("difference a-b:", a.difference(b))
print("symmetric_difference:", a.symmetric_difference(b))
a.add(10)
print("add(10):", a)
a.update([11, 12])
print("update([11, 12]):", a)
a.remove(12)
print("remove(12):", a)
a.discard(100)
print("discard(100):", a)
print("{3,4} is subset of b:", {3, 4}.issubset(b))
print("b is superset of {3,4}:", b.issuperset({3, 4}))
```
