```python
# Inbuilt list methods
nums = [10, 20, 30]
print("Original:", nums)
nums.append(40)
print("append(40):", nums)
nums.extend([50, 60])
print("extend([50, 60]):", nums)
nums.insert(1, 15)
print("insert(1, 15):", nums)
nums.remove(30)
print("remove(30):", nums)
last_item = nums.pop()
print("pop() returned:", last_item, "| now:", nums)
letters = ['b', 'a', 'c', 'a']
print("index('c'):", letters.index('c'))
print("count('a'):", letters.count('a'))
letters.sort()
print("sort():", letters)
letters.reverse()
print("reverse():", letters)
letters_copy = letters.copy()
print("copy():", letters_copy)
letters.clear()
print("clear() -> letters:", letters)
```

```python
# Program 1: remove duplicates while preserving order
data = [1, 2, 2, 3, 1, 4, 4, 5]
unique = []
for x in data:
    if x not in unique:
        unique.append(x)
print("unique in order:", unique)
```

```python
# Program 2: find second largest using sort()
arr = [7, 1, 9, 3, 9, 5]
arr_sorted = sorted(arr)
second_largest = arr_sorted[-2]
print("second largest:", second_largest)
```

```python
# Example: extend() adds multiple elements from another iterable
fruits = ["apple", "banana"]
more_fruits = ["mango", "grapes"]

fruits.extend(more_fruits)
print(fruits)  # ['apple', 'banana', 'mango', 'grapes']
```
