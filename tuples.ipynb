**Key concepts covered:**

- Accessing elements by index
- Tuple unpacking: `x, y = point`

- Practice problems: move zeros, second largest, kth largest element

# Tuples — Immutability & Unpacking

A tuple is like a printed certificate — once it's printed, you cannot change it. A list is like a whiteboard (erasable), but a tuple is permanent ink. Great for data that should never be modified, like coordinates, RGB colors, or dates!

```python
# list  = [1, 2, 3]   ← can change 
# tuple = (1, 2, 3)   ← frozen, cannot change 

#       UNPACKING a tuple:
#       point = (10, 20)
#          ↓      ↓
#          x      y
#       x = 10, y = 20  ← each value goes into its own variable
```

```python
# Creating a tuple — round brackets
point = (10, 20)
colors = ("red", "green", "blue")
single = (42,)             # ← comma needed for single item tuple!

# Accessing — same as lists, use index
print(point[0])            # 10
print(colors[-1])          # blue

# ❌ Tuples are IMMUTABLE — cannot change them!
# point[0] = 99              # TypeError: 'tuple' object does not support item assignment

# Unpacking — assign each value to its own variable
x, y = point
print(x)                   # 10
print(y)                   # 20

# Unpacking works with any iterable!
r, g, b = colors
print(r)                   # red

# Swap two variables elegantly using tuple unpacking
a = 5
b = 10
a, b = b, a                # ✅ no temp variable needed!
print(a, b)                # 10 5

# Tuple of tuples — common in LeetCode!
students = [("Alice", 90), ("Bob", 85), ("Charlie", 92)]
for name, score in students:   # unpack directly in loop!
    print(name, "→", score)
# Alice → 90
# Bob → 85
# Charlie → 92
```

```python
def move_zeros(nums):

 n = len(nums)
 left = 0 
 for right in range(n):
  if nums[right] !=0:
   nums[left] , nums[right] = nums[right], nums[left]
   left += 1
 return nums




```

```python
nums = [ 1, 0 , 3,5,0 , 4]
a = move_zeros(nums)
```

```python
def second_largest(arr):
    largest = float('-inf')
    second = float('-inf')

    for num in arr:
        if num > largest:
            second = largest
            largest = num

        elif num > second and num != largest:
            second = num

    return second


arr = [3, 2, 1, 5, 6, 4]
print(second_largest(arr))
```

```python
def kth_largest(arr, k):
    arr.sort(reverse=True)
    return arr[k-1]


arr = [3, 2, 1, 5, 6, 4]
k = 2

print(kth_largest(arr, k))
```

```python
import heapq

def kth_largest(arr, k):
    heap = []

    for num in arr:
        heapq.heappush(heap, num)

        if len(heap) > k:
            heapq.heappop(heap)

    return heap[0]


arr = [3,2,1,5,6,4]
k = 2

print(kth_largest(arr,k))
```
