# 📖 Python Dictionaries

Dictionaries store data as **key-value** pairs — like a real dictionary where you look up a word (key) to get its definition (value).

**Key concepts covered:**
- Creating dicts with `{key: value}`
- Accessing: `dict[key]` vs safe `dict.get(key, default)`
- Adding and updating key-value pairs
- Looping with `.items()`, `.keys()`, `.values()`
- Membership check: `key in dict`
- Frequency counting pattern (a LeetCode favourite!)

# Dictionaries — key-value, get, update

Real-world analogy:
A dictionary is like a real dictionary book — you look up a word (key) and get its definition (value). Instead of searching page by page, you jump directly to what you need. Super fast! 

<!-- student = {"name": "Alice", "age": 25, "grade": "A"}

```python
# Creating a dictionary — key: value pairs inside {}
student = {"name": "Alice", "age": 25, "grade": "A"}

# Accessing values by key
print(student["name"])         # Alice
print(student["age"])          # 25

# ❌ Accessing a missing key crashes!
# print(student["phone"])      # KeyError!

# ✅ get() — safe lookup, returns None if key missing
print(student.get("phone"))    # None  ← no crash!
print(student.get("phone", "N/A"))  # N/A  ← custom default

# Adding a new key-value pair
student["city"] = "Delhi"
print(student)   # {"name":"Alice","age":25,"grade":"A","city":"Delhi"}

# Updating an existing value
student["grade"] = "A+"
print(student["grade"])        # A+

# Looping through a dictionary
for key, value in student.items():
    print(key, "→", value)
# name → Alice
# age  → 25
# grade → A+
# city → Delhi

# Check if a key exists
print("name" in student)       # True
print("phone" in student)      # False

# Common dictionary methods
print(student.keys())          # dict_keys(['name', 'age', 'grade', 'city'])
print(student.values())        # dict_values(['Alice', 25, 'A+', 'Delhi'])
```

```python
# What to remember:

# Keys must be unique — setting the same key twice just updates the value
# Use get() instead of [] when you're not sure the key exists — avoids crashes!
# Dictionaries are unordered — don't rely on a specific sequence
# This is the #1 structure for frequency counting in LeetCode! 
```

```python
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
```

```python
count = {}
for i in words:
   if i in count:
     count[i] = count[i] + 1
   else:
     count[i] = 1
          
          
```

```python
count
```

```python
count["apple"] = 10
count
```

```python
# Inbuilt dictionary methods
student = {"name": "Nandan", "age": 25, "city": "Delhi"}
print("Original:", student)
print("get('name'):", student.get("name"))
print("get('phone', 'N/A'):", student.get("phone", "N/A"))
print("keys():", list(student.keys()))
print("values():", list(student.values()))
print("items():", list(student.items()))
student.update({"age": 26, "grade": "A"})
print("update(...):", student)
student.setdefault("country", "India")
print("setdefault('country', 'India'):", student)
removed_age = student.pop("age")
print("pop('age') returned:", removed_age, "| now:", student)
last_pair = student.popitem()
print("popitem() returned:", last_pair, "| now:", student)
student_copy = student.copy()
print("copy():", student_copy)
student_copy.clear()
print("clear() ->", student_copy)
```

```python
# Program 1: character frequency counter
word = "mississippi"
char_count = {}
for ch in word:
    char_count[ch] = char_count.get(ch, 0) + 1
print("char frequency:", char_count)
```

```python
# Program 2: invert dictionary (value -> key)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print("inverted dict:", inverted)
```
