```python
def lower_triangular(n):
    for i in range(1, n + 1):
        print('*' * i)

n = 5
lower_triangular(n)
```

```python
def upper_triangular(n):
    for i in range(n, 0, -1):
        print('*' * i)

n = 5
upper_triangular(n)
```

```python
def pyramid(n):
    for i in range(1, n + 1):
        print(' ' * (n - i) + '*' * (2 * i - 1))

n = 5
pyramid(n)
```

```python
# Brute Force Approach
def product_except_self_bruteforce(nums):
    out = []
    for i in range(len(nums)):
        p = 1
        for j in range(len(nums)):
            if i != j:
                p = p * nums[j]
        out.append(p)
    return out

print(product_except_self_bruteforce([1,2,3,4]))
```
