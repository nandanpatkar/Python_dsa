# 🔤 Python String Methods

Strings are immutable sequences of characters. Python provides rich built-in methods to manipulate them.

**Key concepts covered:**
- `upper()` / `lower()` — change case
- `split()` — chop a string into a list
- `join()` — glue a list back into a string
- `replace()` — substitute substrings
- `strip()` / `removeprefix()` / `removesuffix()` — clean whitespace/prefixes

# String Methods — split, join, upper, lower, replace

```python
# sentence = "hello world"

#   upper()          lower()           split()
# ┌───────────┐    ┌───────────┐    ┌─────────────────┐
# │ "HELLO    │    │ "hello    │    │ ["hello","world"]│
# │  WORLD"   │    │  world"   │    │  ← list!        │
# └───────────┘    └───────────┘    └─────────────────┘

#   replace()                join()
# ┌──────────────────┐    ┌──────────────────────┐
# │ "hello python"   │    │ "hello-world"        │
# │ "world"→"python" │    │ "-".join([...])      │
# └──────────────────┘    └──────────────────────┘
```

```python
sentence = "hello world"

# upper() / lower() — change case
print(sentence.upper())          # HELLO WORLD
print("HELLO".lower())           # hello

# split() — chops string into a LIST by a separator
words = sentence.split(" ")      # split on space
print(words)                     # ['hello', 'world']

csv = "apple,banana,cherry"
fruits = csv.split(",")          # split on comma
print(fruits)                    # ['apple', 'banana', 'cherry']

# join() — glues a LIST back into a STRING
words = ["hello", "world"]
print(" ".join(words))           # hello world
print("-".join(words))           # hello-world
print("".join(words))            # helloworld

# replace() — swaps one part for another
print(sentence.replace("world", "python"))  # hello python
print("aabbcc".replace("b", "x"))           # aaxxcc

# strip() — removes whitespace from both ends (bonus!)
messy = "   hello   "
print(messy.strip())             # "hello"
```

```python
# Inbuilt string methods
text = "  hello python world  "
print("Original:", repr(text))
print("strip():", text.strip())
print("upper():", text.upper())
print("lower():", text.lower())
print("title():", text.title())
print("replace('python', 'coding'):", text.replace("python", "coding"))
words = text.strip().split(" ")
print("split():", words)
print("'-'.join(words):", "-".join(words))
print("startswith('  he'):", text.startswith("  he"))
print("endswith('  '):", text.endswith("  "))
print("count('o'):", text.count("o"))
print("find('python'):", text.find("python"))
```

```python
# Program 1: word frequency counter
sentence = "apple banana apple mango banana apple"
freq = {}
for w in sentence.split():
    freq[w] = freq.get(w, 0) + 1
print("word frequency:", freq)
```

```python
# Program 2: reverse each word in a sentence
s = "learn python fast"
rev_words = [w[::-1] for w in s.split()]
print("reversed words:", " ".join(rev_words))
```
