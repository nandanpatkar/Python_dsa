# 🔄 Python Loops

Loops let you execute a block of code repeatedly.

**Key concepts covered:**
- `for` loop — iterate over a sequence
- `range()` — generate numeric sequences (start, stop, step)
- `while` loop — repeat until a condition is False
- `if / elif / else` — conditional logic inside loops

# Concept 4: Loops — for loop, range(), while loop

```python
# fruits = ["apple", "banana", "cherry"]

#   for loop runs 3 times:
#   ┌─────────────────────────────┐
#   │ Round 1 → fruit = "apple"  │
#   │ Round 2 → fruit = "banana" │
#   │ Round 3 → fruit = "cherry" │
#   └─────────────────────────────┘
#         then it stops 
```

```python
fruits = ["apple", "banana", "cherry" , "mango" , "grapes"]
```

```python
for fruit in fruits: ## for loop — go through each item in the list and assigns it to the variable fruit
    print(fruit)
```

```python
for fruit in range(5): ## for loop — go through each item in the list will print the index of the list
    print(fruit)
```

```python
for fruit in range(5): ## for loop — go through each item in the list
    print(fruits)
```

```python
# range(start, stop) — control where to begin
for i in range(1, 4):  # i goes 1, 2, 3 (4 is excluded!)
    print(i)
```

#  while loop

```python
count = 0
while count < 3:
    print(count)       # prints 0, 1, 2
    count = count + 1  # IMPORTANT: always update, or it loops forever!
```

```python
# What to remember:

# for loops are for when you know how many times to loop
# while loops are for when you loop until something changes
# range(5) gives 0,1,2,3,4 — never includes the stop number
# # Always update your while counter or you'll get an infinite loop! 
```

```python
numbers = [3, 7, 1, 9, 4]

for number in numbers:
    print(number)
```

```python
numbers = [3, 7, 1, 9, 4]

for number in numbers:
    if number > 4:
        print(number)
```

```python
age = 20 

if age <= 13:
    print("child")
elif age <= 18:
    print("teenager")
else:
    print("adult")
```

```python
# Inbuilt helpers used with loops
print("range(1, 10, 2):", list(range(1, 10, 2)))

fruits = ["apple", "banana", "mango"]
for i, fruit in enumerate(fruits, start=1):
    print(f"enumerate -> {i}: {fruit}")

names = ["A", "B", "C"]
scores = [90, 75, 88]
for name, score in zip(names, scores):
    print(f"zip -> {name}: {score}")

nums = [2, 4, 6, 8]
print("all even:", all(n % 2 == 0 for n in nums))
print("any > 7:", any(n > 7 for n in nums))
```

```python
# Program 1: multiplication table using for loop + range
n = 5
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```

```python
# Program 2: count vowels using loop
text = "nandan patkar"
vowels = "aeiou"
count = 0
for ch in text.lower():
    if ch in vowels:
        count += 1
print("vowel count:", count)
```
