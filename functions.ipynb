#  Functions — def, return, parameters

Real-world analogy:
A function is like a vending machine. You press a button (call the function), it takes your input (parameters), does its thing inside, and gives you back a result (return value). You can use it over and over without rewriting the same steps!

```python
# INPUT          FUNCTION BOX           OUTPUT
#         ┌─────┐       ┌─────────────┐        ┌─────┐
#    5 →  │     │  →→→  │  add_two()  │  →→→   │  7  │
#    2 →  │     │       │  5 + 2 = 7  │        │     │
#         └─────┘       └─────────────┘        └─────┘
#       parameters         (the work)         return value
```

```python
# def — defines a new function
# add_two — the name you give it
# (a, b) — parameters (inputs it expects)
def add_two(a, b):
    result = a + b
    return result      # sends the answer back to whoever called it

# Calling the function
print(add_two(5, 2))   # 7
print(add_two(10, 3))  # 13  ← reuse it as many times as you want!

# Function with no parameters
def greet():
    print("Hello, coder! 👋")

greet()                # Hello, coder! 👋

# Function with a default parameter
def greet_name(name="stranger"):
    print("Hello, " + name + "!")

greet_name("Alice")    # Hello, Alice!
greet_name()           # Hello, stranger!  ← uses default
```

#  Concept 6b: Scope — Local vs Global Variables


Real-world analogy:
Think of scope like rooms in a house. A variable created inside a room (function) stays in that room — nobody outside can see it. But a variable in the hallway (global) can be seen from every room.

```python
# HALLWAY (global)
# ┌─────────────────────────────────┐
# │  name = "Alice"                 │
# │                                 │
# │  ┌──────────────────────────┐   │
# │  │  ROOM (function)         │   │
# │  │  age = 25  ← local only  │   │
# │  │  can see name ✅          │   │
# │  └──────────────────────────┘   │
# │                                 │
# │  print(age) ❌ ERROR!           │
# └─────────────────────────────────┘
```

```python
# Global variable — lives outside all functions
name = "Alice"

def greet():
    # Local variable — only exists INSIDE this function
    ages = 25
    print(name)   # ✅ can read global variable
    print(ages)    # ✅ works fine inside the function

greet()

```

```python
print(name)       # ✅ works — name is global
print(ages)        # ❌ ERROR! age doesn't exist out here
```

```python
# Modifying a global variable inside a function
count = 0

def increment():
    global count       # tells Python: use the HALLWAY variable
    count = count + 1

increment()
# increment()
print(count)       # 1
```

```python
# Modifying a global variable inside a function
count = 0

def increment():
    global count       # tells Python: use the HALLWAY variable
    count = count + 1

increment()
increment()
print(count)       # 2
```

```python
# What to remember:

# Variables inside a function are local — they disappear when the function ends
# Functions can read globals freely, but to change one, use global keyword
# Avoid overusing global — it makes code messy. Prefer return instead!
```

```python
x = 10

def change():
    x = 99
    print(x)

change()
print(x)
```

# Multiple Return Values

Real-world analogy:
Think of a chef who prepares a meal and hands you back a full plate — main dish, side, and dessert all at once. A function can return multiple things in one go, not just one!

```python
# INPUT            FUNCTION BOX              OUTPUT
#         ┌─────┐       ┌──────────────────┐        ┌──────────────┐
#   10 →  │     │  →→→  │  min_max(nums)   │  →→→   │  min = 1     │
#    →    │     │       │  finds min & max │        │  max = 10    │
#         └─────┘       └──────────────────┘        └──────────────┘
#                                                   two values back
```

```python
numbers = [3, 7, 1, 9, 4]

def min_max(numbers):
    max_num = max(numbers)
    min_num = min(numbers)
    return min_num , max_num # returns two values as a tuple
```

```python
min_num, max_num = min_max(numbers)
min_num, max_num 
```

```python
def add_min_max(min_num, max_num):
    total = min_num + max_num
    return total
```

```python
total = add_min_max(min_num, max_num)
```

```python
total
```

```python
total = add_min_max(10, 20)
```

```python
total
```

```python
# You can return as many values as you need!
def profile(name, age):
    greeting = "Hello, " + name
    is_adult = age >= 18
    return greeting, is_adult, age * 2   # 3 values! tuple!
```

```python
profile("Nandan", 25)  # returns a tuple with 3 values
```

```python
msg, adult, double = profile("Alice", 20)
print(msg)     # Hello, Alice
print(adult)   # True
print(double)  # 40
```

```python
# What to remember:

# Python lets you return multiple values separated by commas
# Under the hood, Python packs them into a tuple (you'll learn this soon!)
# Unpack them into separate variables using a, b = function()
# # Number of variables on the left must match the number of returned values!
```

```python
def circle(radius):
    area = 3.14 * radius ** 2
    circumference = 2 * 3.14 * radius
    return area, circumference
```

```python
circle(5)  # returns area and circumference as a tuple
```

```python
area , circumference = circle(5)
```

```python
area, circumference
```

# *args and **kwargs — Flexible Inputs

Real-world analogy:
Think of *args like a party invitation — you don't know how many people are coming, so you just say "everyone's welcome!" And **kwargs is like a hotel check-in form — guests fill in their own named fields (name, room type, floor).

```python
# *args — any number of positional inputs
# ┌─────────────────────────────────────┐
# │  add(1, 2, 3, 4, 5)                 │
# │       ↓  ↓  ↓  ↓  ↓                │
# │  args = (1, 2, 3, 4, 5)  ← tuple   │
# └─────────────────────────────────────┘

# **kwargs — any number of NAMED inputs
# ┌─────────────────────────────────────┐
# │  info(name="Alice", age=25)         │
# │         ↓              ↓           │
# │  kwargs = {"name":"Alice","age":25} │
# └─────────────────────────────────────┘
```

```python
# *args — accepts ANY number of positional arguments
def add_all(*args):
    print(args)          # (1, 2, 3, 4) ← it's a tuple!
    total = 0
    for num in args:     # loop through all inputs
        total += num     # same as total = total + num
    return total

print(add_all(1, 2))           # 3
print(add_all(1, 2, 3, 4, 5)) # 15  ← works with any count!


# **kwargs — accepts ANY number of NAMED arguments
def show_info(**kwargs):
    print(kwargs)         # {'name': 'Alice', 'age': 25}
    for key, value in kwargs.items():
        print(key, "→", value)

show_info(name="Alice", age=25)
# name → Alice
# age  → 25


# Combining both — order matters: normal, *args, **kwargs
def mixed(greeting, *args, **kwargs):
    print(greeting)        # Hello!
    print(args)            # (1, 2, 3)
    print(kwargs)          # {'city': 'Paris'}

mixed("Hello!", 1, 2, 3, city="Paris")
```

# Lambda Functions — One-Liners

Real-world analogy:
A lambda is like a post-it note function — too small to deserve a full name and definition, just a quick throwawa

```python
# Regular function:               Lambda equivalent:
# ┌───────────────────┐          ┌──────────────────────────┐
# │ def square(x):    │    →→→   │ square = lambda x: x * x │
# │     return x * x  │          └──────────────────────────┘
# └───────────────────┘               same thing, one line!
```

```python
# Regular function
def square(x):
    return x * x

# Exact same thing as a lambda
square = lambda x: x * x

print(square(5))    # 25


# Lambda with two inputs
add = lambda a, b: a + b
print(add(3, 4))    # 7


# Where lambdas SHINE — inside other functions!
numbers = [5, 2, 8, 1, 9, 3]

# sorted() with a custom key using lambda
sorted_nums = sorted(numbers)
print(sorted_nums)             # [1, 2, 3, 5, 8, 9]

# Sort a list of tuples by the SECOND item
pairs = [(1, 9), (3, 2), (5, 5)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)            # [(3, 2), (5, 5), (1, 9)]


# filter() — keep only items where lambda returns True
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)                   # [2, 4, 6]


# map() — apply a function to every item
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)                 # [2, 4, 6, 8, 10, 12]
```

```python
def greater_number(a, b):
    number_1 = lambda x : x * x
    number_2 = lambda y : y * y

    squared_1 = number_1(a)  
    squared_2 = number_2(b)  

    return max(squared_1, squared_2)
```

```python
# Inbuilt helpers with functions

def stats(numbers):
    return {
        "sum": sum(numbers),
        "min": min(numbers),
        "max": max(numbers),
        "rounded_avg": round(sum(numbers) / len(numbers), 2)
    }

print("stats([3, 7, 1, 9]):", stats([3, 7, 1, 9]))
print("abs(-12):", abs(-12))
```

```python
# Program 1: factorial using function
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print("factorial(6):", factorial(6))
```

```python
# Program 2: palindrome checker using function + inbuilt methods
def is_palindrome(s):
    cleaned = s.lower().replace(" ", "")
    return cleaned == cleaned[::-1]

print("is_palindrome('Never odd or even'):", is_palindrome("Never odd or even"))

# Program 3: flexible sum using *args
def total_marks(*marks):
    return sum(marks)

print("total_marks(78, 88, 92):", total_marks(78, 88, 92))
```
