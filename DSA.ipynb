# Two Pointers — Intro

Imagine two people walking along a hallway from opposite ends — one from the left, one from the right. They walk toward each other until they meet in the middle. Instead of checking every possible pair (slow!), two pointers work together to find the answer in one pass! 

````python
# array = [1, 2, 3, 4, 5, 6]
#          ↑                ↑
#         LEFT            RIGHT
        
#   Step 1: check left + right → move one pointer
#   Step 2: keep going until left and right meet
  
#          [1, 2, 3, 4, 5, 6]
#           L  →          ← R    getting closer...
#               L  →   ← R
#                   L=R           STOP! ✅
````

````python
def two_sum(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left < right:
        current_sum = arr[left] + arr[right]
        
        if current_sum == target:
            return (arr[left], arr[right])  # Found the pair!
        elif current_sum < target:
            left += 1  # Move left pointer to the right
        else:
            right -= 1  # Move right pointer to the left
            
    return None  # No pair found
````

````python
arr = [1, 2, 3, 4, 5, 6]
target = 7
result = two_sum(arr, target)
result
````

````python
# PROBLEM: Does any pair in this SORTED list add up to target?
# Brute force → O(n²)  Two pointers → O(n) ⚡

# ❌ Brute force — check every possible pair
def two_sum_brute(arr, target):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):   # nested loop = slow!
            if arr[i] + arr[j] == target:
                return [i, j]
    return []
````

````python
# **Visual of two pointers in action:**
# ```
# arr = [1, 2, 3, 4, 6],  target = 6

# Step 1: L=0, R=4 → 1+6=7 > 6 → move R left
# Step 2: L=0, R=3 → 1+4=5 < 6 → move L right
# Step 3: L=1, R=3 → 2+4=6 = 6 ✅ FOUND!
````

````python
def twosum(arr , target):
    left = 0 
    right = len(arr)-1

    while left < right:
        current_sum = arr[left]+arr[right]

        if current_sum == target:
            return (arr[left], arr[right])
        elif current_sum < target:
            left += 1
        else: 
            right -= 1 
    return None 
````

````python
# Brute Force Approach
def merge_k_sorted_lists_bruteforce(lists):
    arr = []
    for l in lists:
        arr.extend(l)
    arr.sort()
    return arr

print(merge_k_sorted_lists_bruteforce([[1,4,5],[1,3,4],[2,6]]))
````
